## 2. Aggregated mean calculation of the order_number column grouped by department_id

## 4. Creating a loyalty flag

Criteria:
- If the maximum orders the user has made is over 40,
  then the customer will be labeled a "Loyal customer."
- If the maximum orders the user has made is over 10 but less than or equal to 40, 
then the customer will be labeled a "Regular customer."
- If the maximum orders the user has made is less than or equal to 10, 
then the customer will be labeled a "New customer."

```python
# Aggregating data with the transform() function
ords_prods_merge ['max_order'] = ords_prods_merge.groupby(['user_id']) ['order_number'].transform(np.max)
```

```python
# Checking the output
ords_prods_merge.head(15)
```

The column max_order is created. It is the max number of orders made by each user ID.

```python
# Creating a 'loyalty' flag using the loc() function
ords_prods_merge.loc [ords_prods_merge ['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc [(ords_prods_merge ['max_order'] <= 40) & (ords_prods_merge ['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc [ords_prods_merge ['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Printing the frequency of the loyalty flags:
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking the updated output
ords_prods_merge.head(15)
```

## 5. Spending habits by loyalty category

```python
# Determining basic statistics of spending habits by loyalty category:
# Loyal custumer, Regular customer, New customer

ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

Loyal customers tend to purchase lower-cost items, with an average price of 10.39, while new customers spend more, with a mean cost of 13.29. Regular customers fall in the middle, averaging 12.50 per item. 
All groups share a minimum price of 1.0, but the extremely high maximum cost of 99,999 suggests possible outliers or data errors.

## 6. Creating a spending flag for each user

Criteria:
- If the mean of the prices of products purchased by a user is lower than 10,
  then flag them as a "Low spender."
- If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a "High spender."

```python
# Aggregating data with the transform() function
ords_prods_merge ['avg_product_price'] = ords_prods_merge.groupby(['user_id']) ['prices'].transform(np.mean)
```

```python
# Checking relevant columns 
ords_prods_merge[['user_id', 'avg_product_price']].head(15)
```

```python
# Creating a spending flag using the loc() function
ords_prods_merge.loc [ords_prods_merge ['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc [ords_prods_merge ['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Printing the frequency of the spending flags:
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Checking relevant columns 
ords_prods_merge[['user_id', 'avg_product_price', 'spending_flag']].head(15)
```

## 7. Creating an order frequency flag for each user

Criteria:

- If the median of "days_since_prior_order" is higher than 20, then the customer should be labeled a "Non-frequent customer."

- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a "Regular customer."

- If the median is lower than or equal to 10, then the customer should be labeled a "Frequent customer."

```python
# Aggregating data with the transform() function
ords_prods_merge ['median_days_between_orders'] = ords_prods_merge.groupby(['user_id']) ['days_since_prior_order'].transform(np.median)
```

```python
# Checking relevant columns 
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(15)
```

```python
# # Creating a order frequency flag using the loc() function
ords_prods_merge.loc [ords_prods_merge ['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc [(ords_prods_merge ['median_days_between_orders'] > 10) & (ords_prods_merge ['median_days_between_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc [ords_prods_merge ['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Printing the frequency of the order frequency flags:
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Checking relevant columns 
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency_flag']].head(15)
```
