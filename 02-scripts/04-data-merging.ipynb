## 2. Merging orders and products dataframes

#### Observation: the most suitable way to merge the two dataframes is to use the product_id column. An inner join would be the best course of action since the information that is presented wouldn't aid in a better understanding in both dataframes.

```python
# Test the merge without overwriting:
pd.merge(orders_products_combined, df_products_checked, on = ['product_id'], indicator = 'Match')
```

####  Observation: I couldn't use indicator=True in the merge test above because the orders_products_combined DataFrame already contains a _merge column. Instead, I used indicator='Match'.

```python
# Drop the merge column in the orders_products_combined df
orders_products_combined = orders_products_combined.drop(['_merge'], axis=1)
```

```python
#check the output 
orders_products_combined.head()
```

```python
# Merge updated DataFrame (permanently)
df_orders_products_combined = df_products_checked.merge(orders_products_combined, on = 'product_id', indicator = True)
```

```python
# Check the output
df_orders_products_combined.head()
```

```python
# Check the shape
df_orders_products_combined.shape
```

```python
# Checking the result of merge using the merge flag
df_orders_products_combined['_merge'].value_counts()
```

```python
# Export data to pkl
df_orders_products_combined.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_merge.pkl'))
```
