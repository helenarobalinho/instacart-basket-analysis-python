## 3. Data Consistency Checks - Customers data

Mixed-type data

```python
# Checking whether the dataframe contains any mixed-type columns:

for col in customers.columns.tolist():
    weird = (customers[[col]].map(type) != customers
             [[col]].iloc[0].apply(type)).any(axis = 1)
    if len(customers[weird]) > 0:
        print(col)
```

```python
# Fixing the mixed-type data in the 'first_name' column:

customers['first_name'] = customers ['first_name'].astype('str')
```

```python
# Checking the output:

customers['first_name'].dtype
```

```python
# A second check:

for col in customers.columns.tolist():
    weird = (customers[[col]].map(type) != customers
             [[col]].iloc[0].apply(type)).any(axis = 1)
    if len(customers[weird]) > 0:
        print(col)
```

```python
# All good now.
```

## 4. Combining the Customers data with the rest of the prepared Instacart data

```python
# Importing the prepared dataframe 'orders_products_merged_grouped'

ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_grouped.pkl'))
```

```python
# Checking the output

ords_prods_merge.head()
```

```python
# Checking the size

ords_prods_merge.shape
```

Key-column to merge the 2 dataframes: 'user_id'

```python
# Making sure the data type matches the one of the customers df

ords_prods_merge['user_id'].dtype
```

```python
# Converting the data type to string:

ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('str')
```

```python
# Checking results:

ords_prods_merge['user_id'].dtype
```

Checking unique values in the user_id column in both dataframes

```python
# Checking unique values in the user_id column (ords_prods_merge)

ords_prods_merge['user_id'].nunique()
```

```python
# Checking unique values in the user_id column (customers)

customers['user_id'].nunique()
```

We have the same amount of unique values and the same data type, we can now merge both dataframes by the shared key_column.

```python
# Merging test: the 'outer' merge will combine all the observations and show the real merge rate

merge_test = pd.merge (ords_prods_merge, customers, on = 'user_id', how = 'outer', indicator = 'merging_status')
```

```python
# Let's verify the merging results:

merge_test ['merging_status'].value_counts()
```

```python
# The outcome is a full match, so we can move to a permanent merge:

final_merged = ords_prods_merge.merge(customers, on = 'user_id', how = 'inner')
```

```python
# Checking the output

final_merged.head()
```

```python
# Checking the size:

final_merged.shape
```
