### 2.1 Mixed Data Types

```python
# Create a dataframe (for simulation purposes)
df_test = pd.DataFrame ()
```

```python
# Create a mixed type column
df_test ['mix'] = ['a', 'b', 1, True]
```

```python
df_test.head()
```

```python
# Function to check if the df contains mixed-type columns (applymap foi substituido por map)

for col in df_test.columns.tolist ():
    weird = (df_test [[col]].map(type) != df_test [[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_test [weird]) > 0:
        print (col)
```

```python
# Fix the mixed data types by forcing ot to be all string
df_test['mix'] = df_test['mix'].astype('str')
```

### 2.2 Missing Values

```python
# Finding the missing values
df_products.isnull(). sum()
```

```python
# Viewing the missing data
df_nan = df_products[df_products ['product_name'].isnull() == True]
```

```python
df_nan
```

- The missing values are small and cannot be guessed, the product_name is a string - Let's drop it.

```python
# Since overwriting is risky, we can create a subset dataframe:

df_products_clean = df_products[df_products['product_name'].isnull() == False]
```

```python
# Let's compare. Before:
df_products.shape
```

```python
# After:
df_products_clean.shape
```

### Exactly 16 rows were dropped from the original df ✅ 

### 2.3 Duplicates

```python
# Creating a subset of the duplicate rows to find them:
df_dups = df_products_clean[df_products_clean.duplicated()]
```

```python
df_dups
```

```python
# Removing the duplicates by creating a new version of the df without them:
df_products_clean_no_dups = df_products_clean.drop_duplicates()
```

```python
# Let's compare it. Before:
df_products_clean.shape
```

```python
# After:
df_products_clean_no_dups.shape
```
